```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Define the make_graph function
def make_graph(stock_data, revenue_data, stock_name, revenue_col_name):
    fig, ax1 = plt.subplots(figsize=(14, 7))

    # Plot stock data
    color = 'tab:red'
    ax1.set_xlabel('Date')
    ax1.set_ylabel('Stock Price', color=color)
    ax1.plot(stock_data.index, stock_data['Close'], color=color)
    ax1.tick_params(axis='y', labelcolor=color)
    
    # Plot revenue data if columns are present
    if 'Date' in revenue_data.columns and revenue_col_name in revenue_data.columns:
        ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
        color = 'tab:blue'
        ax2.set_ylabel('Revenue', color=color)
        ax2.plot(revenue_data['Date'], revenue_data[revenue_col_name], color=color)
        ax2.tick_params(axis='y', labelcolor=color)
    else:
        print(f"One or more required columns are missing. Cannot plot revenue data.")
    
    fig.tight_layout()  # otherwise the right y-label is slightly clipped
    plt.title(f"{stock_name} Stock Price and Revenue")
    plt.show()

# Download GameStop stock data
gme_data = yf.download('GME', start='2010-01-01', end='2023-12-31')

# Load GameStop revenue data
gme_revenue = pd.read_csv('/resources/PY0220EN/labs/project/gme_data.csv')

# Ensure the 'Date' column is in datetime format
gme_revenue['Date'] = pd.to_datetime(gme_revenue['Date'])

# Sort by date
gme_revenue = gme_revenue.sort_values(by='Date')

# Display the column names to debug
print("Columns in gme_revenue DataFrame:")
print(gme_revenue.columns)

# Plot the data
# Update 'Revenue' to the correct column name found in the previous step
make_graph(gme_data, gme_revenue, 'GameStop', 'Revenue')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Tesla stock data from CSV
tesla_data = pd.read_csv('tesla_stock_data.csv', parse_dates=['Date'], index_col='Date')

# Function to create the plot
def make_graph(stock_data, title):
    plt.figure(figsize=(12, 6))
    
    # Plot the closing price
    plt.plot(stock_data.index, stock_data['Close'], label='Closing Price', color='tab:blue')
    
    # Add titles and labels
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Price (USD)')
    plt.legend()
    
    # Display the plot
    plt.grid(True)
    plt.show()

# Call the function with Tesla data
make_graph(tesla_data, 'Tesla Stock Closing Prices')
```
