Summary of Charts and Their Purpose
Bar Charts

Purpose: To compare the number of Thai restaurants across different neighborhoods.
Reason: Bar charts are ideal for comparing discrete categories and showing variations in the number of Thai restaurants, which helps identify neighborhoods with high or low competition.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data
neighborhoods = ['Neighborhood A', 'Neighborhood B', 'Neighborhood C']
thai_restaurant_counts = [5, 15, 10]

# Plotting the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=neighborhoods, y=thai_restaurant_counts, palette='viridis')
plt.title('Number of Thai Restaurants in Different Neighborhoods')
plt.xlabel('Neighborhoods')
plt.ylabel('Number of Thai Restaurants')
plt.show()
```

Pie Charts

Purpose: To illustrate the distribution of various types of venues within the neighborhoods.
Reason: Pie charts provide a visual representation of the proportion of different venue types, which helps in understanding the neighborhood's culinary diversity.


```python
# Sample data
venue_types = ['Thai', 'Chinese', 'Japanese', 'Other']
venue_counts = [15, 25, 10, 50]

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(venue_counts, labels=venue_types, autopct='%1.1f%%', colors=sns.color_palette('viridis'))
plt.title('Distribution of Venue Types in Neighborhood A')
plt.show()
```

Scatter Plots

Purpose: To show the relationship between the number of Thai restaurants and other variables such as population density.
Reason: Scatter plots are effective for visualizing correlations between two continuous variables, helping to understand factors that might influence the success of a new restaurant


```python
# Sample data
population_density = [2000, 3000, 2500]
thai_restaurant_counts = [5, 15, 10]

# Plotting the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(population_density, thai_restaurant_counts, c='blue', alpha=0.5)
plt.title('Relationship between Population Density and Number of Thai Restaurants')
plt.xlabel('Population Density (per sq km)')
plt.ylabel('Number of Thai Restaurants')
plt.show()
```

Exploratory Data Analysis


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv")

# Display the first few rows of the dataset
print(df.head())

# Get basic information about the dataset
print(df.info())

# Summary statistics of numerical features
print(df.describe())

# Check for missing values
print(df.isnull().sum())

# Distribution of a categorical feature
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='category_column', palette='viridis')  # Replace 'category_column' with your column name
plt.title('Distribution of Categories')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()

# Scatter plot example
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='x_column', y='y_column', hue='category_column', palette='viridis')  # Replace columns with your actual column names
plt.title('Scatter Plot of X vs Y')
plt.xlabel('X Column')
plt.ylabel('Y Column')
plt.show()

# Correlation heatmap
plt.figure(figsize=(10, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()

# Visualizing missing values using a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```
